```python
import numpy as np
import pandas as pd
from itertools import chain
import IPython.display
import matplotlib.pyplot as plt
```

```python
def open_files(path):
    time_list = []
    time_path = path[:54] +'time_list.txt'
    with open(time_path, 'r') as file:
        for line in file:
            time_list.append(line.strip())
        print(time_list)
    df=pd.read_csv(path)
    display(df.tail())
    return df, time_list

#df1= dataframe from 04.09.23, df2 from 11.09 and df3 from 25.09 
df1, times1 = open_files('/home/finn/visual_Studio_Code/data/2023-09-04/results/results_2023-09-04_HIP_100587_B (Johnson)_30')
df2, times2 = open_files('/home/finn/visual_Studio_Code/data/2023-09-11/results/results_2023-09-11_HIP100587_focused_B (Johnson)_40')
df3, times3 = open_files('/home/finn/visual_Studio_Code/data/2023-09-25/results/results_2023-09-25_HIP100587_B (Johnson)_40')
```

```python
# Get unique 'DR3Name' values from each DataFrame
unique_names_df1 = df1['DR3Name'].unique()
unique_names_df2 = df2['DR3Name'].unique()
unique_names_df3 = df3['DR3Name'].unique()

# Find rows with same 'DR3Name' values across all DataFrames
common_names = set(unique_names_df1) & set(unique_names_df2) & set(unique_names_df3)

# Filter each DataFrame to keep only rows with 'DR3Name' present in common_names
fil_df1 = df1[df1['DR3Name'].isin(common_names)].copy()
fil_df2 = df2[df2['DR3Name'].isin(common_names)].copy()
fil_df3 = df3[df3['DR3Name'].isin(common_names)].copy()

# calc absolute magnitudes for next step
fil_df1['absolute Gmag'] = fil_df1['Gmag'] + 5*np.log10(fil_df1['Plx']/100)
fil_df2['absolute Gmag'] = fil_df2['Gmag'] + 5*np.log10(fil_df2['Plx']/100)
fil_df3['absolute Gmag'] = fil_df3['Gmag'] + 5*np.log10(fil_df3['Plx']/100)

print(len(fil_df1), len(fil_df2), len(fil_df3))
pd.set_option('display.max_rows', None)  # Set to None to display all rows
pd.set_option('display.max_columns', None) 
display(fil_df1)
display(fil_df2)
display(fil_df3)
```

```python
def get_mag_lists(df):
    df_ft = df.iloc[:-1,:] # all rows expept last one (bright star), all columns
    #display(df_ft)
    df_br = df.iloc[-1:,:] #just last row
    #display(df_br)

    mag_mean_columns_ft = [col for col in df_ft.columns if 'Mag_Mean_Im' in col]
    mag_mean_ft = df_ft[mag_mean_columns_ft].values.T.tolist()
    mean_mag_mean_ft = [np.mean(sublist) for sublist in mag_mean_ft]

    mag_median_columns_ft = [col for col in df_ft.columns if 'Mag_Median_Im' in col]
    mag_median_ft = df_ft[mag_median_columns_ft].values.T.tolist()
    mean_mag_median_ft = [np.mean(sublist) for sublist in mag_median_ft]

    mag_mean_columns_br = [col for col in df_br.columns if 'Mag_Mean_Im' in col]
    mag_mean_br = df_br[mag_mean_columns_br].values.T.tolist() 
    mag_mean_br = list(chain.from_iterable(mag_mean_br)) #to remove inner brackets i.e. [[.],[.],[.]] -> [.,.,.]

    mag_median_columns_br = [col for col in df_br.columns if 'Mag_Median_Im' in col]
    mag_median_br = df_br[mag_median_columns_br].values.T.tolist()
    mag_median_br = list(chain.from_iterable(mag_median_br))

    print(mean_mag_mean_ft)
    print(mean_mag_median_ft)
    print(mag_mean_br)
    print(mag_median_br)
    print('')
    return mean_mag_mean_ft, mean_mag_median_ft, mag_mean_br, mag_median_br

mean_mag_mean_ft1, mean_mag_median_ft1, mag_mean_br1, mag_median_br1 = get_mag_lists(fil_df1)
mean_mag_mean_ft2, mean_mag_median_ft2, mag_mean_br2, mag_median_br2 = get_mag_lists(fil_df2)
mean_mag_mean_ft3, mean_mag_median_ft3, mag_mean_br3, mag_median_br3 = get_mag_lists(fil_df3)
```

```python
from datetime import datetime

def shorten_time(time_list):
    datetime_objects = [datetime.strptime(ts, '%Y-%m-%dT%H:%M:%S') for ts in time_list]
    # Shortened representation of timestamps
    shortened_timestamps = [dt.strftime('%m-%d %H:%M') for dt in datetime_objects]
    return shortened_timestamps

times1_short = shorten_time(times1)
times2_short = shorten_time(times2)
times3_short = shorten_time(times3)

fig = plt.figure(figsize=(20,10))
plt.title('Time-series differential magnitudes of HIP100587')
plt.xlabel('Time [month-day hour:minute]')
plt.ylabel('Differential magnitudes [mag]')
for times, diff_mags in zip([times1_short, times2_short, times3_short[1:]], [diff_mags_mean1, diff_mags_mean2, diff_mags_mean3]):
    plt.errorbar(times, diff_mags, yerr=np.std(diff_mags), linestyle='none',marker='o', label='values for mean stacked series from 23-'+ times[0][:5])
    plt.legend()

for times, diff_mags in zip([times1_short, times2_short, times3_short[1:]], [diff_mags_median1, diff_mags_median2, diff_mags_median3]):
    plt.errorbar(times, diff_mags, yerr=np.std(diff_mags), linestyle='none',marker='o', label='values for median stacked series from 23-'+ times[0][:5])
    plt.legend()
```
