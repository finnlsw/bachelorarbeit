```python
def filter_common_names(df_list):
    if 'DR3Name' not in df_list[0].columns: #check if df_list has gaia names in it
        filtered_df = df_list.copy()
        print('no gaia names in dfs')
        return filtered_df
    else: 
        if isinstance(df_list, pd.DataFrame):  # Check if df_list is a single DataFrame
            filtered_df = df_list.copy()
            filtered_df['absolute Gmag'] = filtered_df['Gmag'] + 5 * np.log10(filtered_df['Plx'] / 100)
            print('success for single frame')
            return filtered_df
        else:  # For multiple DataFrames in the list
            unique_names_list = [df['DR3Name'].unique() for df in df_list]
            common_names = set(unique_names_list[0]).intersection(*unique_names_list[1:])
            filtered_dfs = [df[df['DR3Name'].isin(common_names)].copy() for df in df_list]
            for df in filtered_dfs:
                df['absolute Gmag'] = df['Gmag'] + 5 * np.log10(df['Plx'] / 100)
            print('success for multiple frames')
            return filtered_dfs


fil_dfs = filter_common_names(df_list)
for i, df in enumerate(fil_dfs):
    print(f"Filtered DataFrame {i+1}:")
    display(df)
```

```python
def get_mag_lists(df):
    df_ft = df.iloc[:-1,:] # all rows expept last one (bright star), all columns
    #display(df_ft)
    df_br = df.iloc[-1:,:] #just last row
    #display(df_br)

    mag_mean_columns_ft = [col for col in df_ft.columns if 'Mag_Mean_Im' in col]
    mag_mean_ft = df_ft[mag_mean_columns_ft].values.T.tolist()
    mean_mag_mean_ft = [np.mean(sublist) for sublist in mag_mean_ft]

    mag_median_columns_ft = [col for col in df_ft.columns if 'Mag_Median_Im' in col]
    mag_median_ft = df_ft[mag_median_columns_ft].values.T.tolist()
    mean_mag_median_ft = [np.mean(sublist) for sublist in mag_median_ft]

    mag_mean_columns_br = [col for col in df_br.columns if 'Mag_Mean_Im' in col]
    mag_mean_br = df_br[mag_mean_columns_br].values.T.tolist() 
    mag_mean_br = list(chain.from_iterable(mag_mean_br)) #to remove inner brackets i.e. [[.],[.],[.]] -> [.,.,.]

    mag_median_columns_br = [col for col in df_br.columns if 'Mag_Median_Im' in col]
    mag_median_br = df_br[mag_median_columns_br].values.T.tolist()
    mag_median_br = list(chain.from_iterable(mag_median_br))

    print(mean_mag_mean_ft)
    print(mean_mag_median_ft)
    print(mag_mean_br)
    print(mag_median_br)
    print('')
    return mean_mag_mean_ft, mean_mag_median_ft, mag_mean_br, mag_median_br



def get_mag_lists_mesh(df_list):
    df_ft = df_list[0] # all rows expept last one (bright star), all columns
    #display(df_ft)
    df_br = df_list[1] #just last row
    #display(df_br)

    mag_mean_columns_ft = [col for col in df_ft.columns if 'Mag_Mean_Im' in col]
    mag_mean_ft = df_ft[mag_mean_columns_ft].values.T.tolist()
    mean_mag_mean_ft = [np.mean(sublist) for sublist in mag_mean_ft]

    mag_median_columns_ft = [col for col in df_ft.columns if 'Mag_Median_Im' in col]
    mag_median_ft = df_ft[mag_median_columns_ft].values.T.tolist()
    mean_mag_median_ft = [np.mean(sublist) for sublist in mag_median_ft]

    mag_mean_columns_br = [col for col in df_br.columns if 'Mag_Mean_Im' in col]
    mag_mean_br = df_br[mag_mean_columns_br].values.T.tolist() 
    #mag_mean_br = list(chain.from_iterable(mag_mean_br)) #to remove inner brackets i.e. [[.],[.],[.]] -> [.,.,.]
    mag_mean_br = [np.mean(sublist) for sublist in mag_mean_br]

    mag_median_columns_br = [col for col in df_br.columns if 'Mag_Median_Im' in col]
    mag_median_br = df_br[mag_median_columns_br].values.T.tolist()
    #mag_median_br = list(chain.from_iterable(mag_median_br))
    mag_median_br = [np.mean(sublist) for sublist in mag_median_br]

    print(mean_mag_mean_ft)
    print(mean_mag_median_ft)
    print(mag_mean_br)
    print(mag_median_br)
    print('')
    return mean_mag_mean_ft, mean_mag_median_ft, mag_mean_br, mag_median_br


mean_mag_mean_ft_list = []
mean_mag_median_ft_list = []
mag_mean_br_list = []
mag_median_br_list = []

if series_type != 'mesh ':
    print('no mesh')
    for df in fil_dfs:
        mean_mag_mean_ft, mean_mag_median_ft, mag_mean_br, mag_median_br = get_mag_lists(df)
        mean_mag_mean_ft_list.append(mean_mag_mean_ft)
        mean_mag_median_ft_list.append(mean_mag_median_ft)
        mag_mean_br_list.append(mag_mean_br)
        mag_median_br_list.append(mag_median_br)

else: 
    print('mesh')
    mean_mag_mean_ft, mean_mag_median_ft, mag_mean_br, mag_median_br = get_mag_lists_mesh(df_list[:2]) #chante later back to (df_list) !!!!!!!!
    mean_mag_mean_ft_list.append(mean_mag_mean_ft)
    mean_mag_median_ft_list.append(mean_mag_median_ft)
    mag_mean_br_list.append(mag_mean_br)
    mag_median_br_list.append(mag_median_br)
    mean_mag_mean_ft, mean_mag_median_ft, mag_mean_br, mag_median_br = get_mag_lists_mesh(df_list[2:])
    mean_mag_mean_ft_list.append(mean_mag_mean_ft)
    mean_mag_median_ft_list.append(mean_mag_median_ft)
    mag_mean_br_list.append(mag_mean_br)
    mag_median_br_list.append(mag_median_br)
```

```python
def calc_results(mag_ft_mean, mag_br_mean, mag_ft_median, mag_br_median):
    diff_mags_mean = [m_br - m_ft for m_br, m_ft in zip(mag_br_mean, mag_ft_mean)]
    diff_mags_median = [m_br - m_ft for m_br, m_ft in zip(mag_br_median, mag_ft_median)]
    print(f'The diff Mag (mean stacked) is:   M = ${round(np.mean(diff_mags_mean), 5)}\pm{round(np.std(diff_mags_mean), 5)}$')
    print(f'The diff Mag (median stacked) is: M = ${round(np.mean(diff_mags_median), 5)}\pm{round(np.std(diff_mags_median), 5)}$')
    print('')
    print(f'The std for target (mean stacked):   {round(np.std(mag_br_mean), 5)} and for the bkg stars (mean values):    {round(np.std(mag_ft_mean), 5)}')
    print(f'The std for target (median stacked): {round(np.std(mag_br_median), 5)} and for the bkg stars (median stacked): {round(np.std(mag_ft_median), 5)}')
    print('')
    print('')
    return diff_mags_mean, diff_mags_median


diff_mags_mean_list = []
diff_mags_median_list = []
for i in range(len(mean_mag_mean_ft_list)):
    print(f'for df{i+1}:')
    diff_mags_mean, diff_mags_median = calc_results(mean_mag_mean_ft_list[i], mag_mean_br_list[i], mean_mag_median_ft_list[i], mag_median_br_list[i])                                      
    diff_mags_mean_list.append(diff_mags_mean)
    diff_mags_median_list.append(diff_mags_median)
```

```python
def plot_ts(input_list, title, plot_mean=False):
    
    length = len(short_times) 
    fig, ax = plt.subplots(1, len(short_times), sharey=True, facecolor='w', figsize = (length*6,10))
    ax[length//2].set_title(title)
    ax[0].set_ylabel('Magnitudes [mag]')
    for i in range(length):
        for j in range(length):
            ax[j].errorbar(short_times[i], input_list[i], yerr=np.std(input_list[i]), linestyle='none', marker='o', 
                            label='series from 23-' + short_times[i][0][:5], capsize=2)
        ax[i].tick_params(axis='x', rotation=45)
        ax[i].set_xlabel('Time [month-day hour:minute]')

        buffer = 0.5
        ax[i].set_xlim(short_times[i][0], short_times[i][-1])
        ax[i].set_xlim(ax[i].get_xlim()[0] - buffer, ax[i].get_xlim()[1] + buffer)


        if i > 0:
            ax[i].spines['left'].set_visible(False)
            ax[i].tick_params(left=False, right=False)  # Remove ticks in the middle subplot

        d = .015  # how big to make the diagonal lines in axes coordinates  
        if i < length - 1:
            ax[i].spines['right'].set_visible(False)
            kwargs = dict(transform=ax[i].transAxes, color='k', clip_on=False)
            ax[i].plot((1-d, 1+d), (-d, +d), **kwargs)        # top-left diagonal
            #ax[i].plot((1-d, 1+d), (1-d, 1+d), **kwargs)
            kwargs.update(transform=ax[i+1].transAxes)
            #ax[i+1].plot((-d, +d), (1-d, 1+d), **kwargs)  # bottom-left diagonal
            ax[i+1].plot(( - d, + d), (- d,  + d), **kwargs)        # bottom-left diagonal

        if plot_mean == True:
            means_mean_stacked = [np.mean(diff_mags) for diff_mags in input_list]
            print(f'The mean of all {length} series is: {round(np.mean(means_mean_stacked),4)} with std. of: {round(np.std(means_mean_stacked),4)}')
            ax[i].axhline(np.mean(means_mean_stacked), label=f'mean of all series', color='black', linestyle='--', alpha=0.5)
    
    ax[length-1].legend()
    fig.subplots_adjust(hspace=0.5)
    plt.tight_layout()


plot_ts(diff_mags_mean_list, title=f'Mean stacked time series for the differential magnitudes of {target} for {series_type} series in filter {filter_list[0]}'  ,  plot_mean=True)
```
